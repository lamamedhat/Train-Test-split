# Train Test Split 
The train-test split is a technique for evaluating the performance of a machine learning algorithm.

```python
import pandas as pd

data = pd.read_csv('G:\\datasets\\diabetics.csv')

print (data.shape)  # "shape" show number of raws and columns 
print (data.head)

# axis=1 -> column
# axis=0 -> row

x=data.drop("outcome",axis=1)
y=data['outcome']
from sklearn.model_selection import train_test_split

# Split dataset into training set and test set

# X=feature
# y=label (target)
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size = 0.3) # 70% training and 30% test

#  describe -> mean , std , min , max .... 
print (data['outcome'].describe())

print (data.ndim) # number of dimentions of matrix

from sklearn import svm


#Create a svm Classifier
clf = svm.SVC(kernel='linear') # Linear Kernel

#Train the model using the training sets
clf.fit(X_train, y_train)

#Predict the response for test dataset
y_pred = clf.predict(X_test)

# confusion matrix

from sklearn.metrics import confusion_matrix

conf = confusion_matrix(y_pred,y_test)

print('confusion matrix',conf)



#Import scikit-learn metrics module for accuracy calculation
from sklearn import metrics

# Model Accuracy: how often is the classifier correct?

print("Accuracy:",metrics.accuracy_score(y_test, y_pred))

print("precition:",metrics.precision_score(y_test, y_pred))

print("f1 score:",metrics.f1_score(y_test, y_pred))

print("recall:",metrics.recall_score(y_test, y_pred))


```
